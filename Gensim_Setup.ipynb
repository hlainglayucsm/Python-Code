```python
#Custom Embeddings
from gensim.models import Word2Vec
# define training data
sentences = [['this', 'is', 'the', 'first', 'sentence', 'for', 'word2vec'],
			['this', 'is', 'the', 'second', 'sentence'],
			['yet', 'another', 'sentence'],
			['one', 'more', 'sentence'],
			['and', 'the', 'final', 'sentence']]
# train model
model = Word2Vec(sentences, min_count=1)
# summarize the loaded model
print(model)
# summarize vocabulary
words = list(model.wv.vocab)
print(words)
# access vector for one word
print(model['sentence'])
# save model
model.save('custom_model.bin')
# load model
new_model = Word2Vec.load('custom_model.bin')
print(new_model)
similar_word=new_model.wv.most_similar('sentence')
print("Similar word of sentence>>>>",similar_word)
```

```python
#Word Embeddings with Corpus
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
lineLst=list()
tokenLst=list()
senLst=list()
total_token_lst=list()
with open("TestCorpus.txt") as file:
    for line in file:
        lineLst.append(line)
for eachLine in lineLst:
    print("Each Line===>",eachLine)
    print("===============================================================")
print("Length==>", len(lineLst))
print("===============================================================")
for l in lineLst:
    senLst.append(sent_tokenize(l))
print( "Sentence Lst::::",senLst)
print("===============================================================")
for each_chuck in senLst:
    print("each word lst===",each_chuck)
    print("===============================================================")
    for sen in each_chuck:
        print("each sen==>",sen)
        tokenLst.append(word_tokenize(sen))
    print("===============================================================")
print("Tokenized word Lst===>",tokenLst)
print("length of token lst==>",len(tokenLst))
print("===============================================================")
```

```python
from gensim.models import Word2Vec
word2vec = Word2Vec(tokenLst, min_count=1)
vocabulary = list(word2vec.wv.vocab)
print("word2vec====",word2vec)
print("list===>",vocabulary)
print("weather==>",word2vec['weather'])
word2vec.save('eng_wordEmbedding_model.bin')
new_word2Vec=Word2Vec.load('eng_wordEmbedding_model.bin')
print("__new word2vec___",new_word2Vec)
```
