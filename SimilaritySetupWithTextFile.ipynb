```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
file=open('word_pair.txt',"r")
first_column_lst=[]
second_column_lst=[]
second_column_clean_lst=[]
min_value_lst=[]
max_value_lst=[]
word_pair_lst=file.readlines()
print("word pair lst===",word_pair_lst)
for pair in word_pair_lst:
    pair_split_lst=pair.split("\t")
    first_column_lst.append(pair_split_lst[0])
    second_column_lst.append(pair_split_lst[1])
    
print("first list==",first_column_lst)
print("second list==",second_column_lst)
print("=================================================")

# Clean \n in the second word "bride\n" to get only word "bride"
for second_word in second_column_lst:
    second_column_clean_lst.append(second_word.split("\n")[0])
print("clean data==",second_column_clean_lst)

for index,first_word in enumerate(first_column_lst):
    first_word_synsetLst=wn.synsets(first_word) # synset of first word 
    second_word=second_column_clean_lst[index]
    second_word_synsetLst=wn.synsets(second_word) # synset of second word
    print("first synset list==",first_word_synsetLst)
    print("second synset list==",second_word_synsetLst)
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
    similarity_lst=[]
    for first_synsetWord in first_word_synsetLst:       
        for second_synsetWord in second_word_synsetLst:
            each_similarity=first_synsetWord.path_similarity(second_synsetWord)
            similarity_lst.append(each_similarity)
            print(first_synsetWord," and ",second_synsetWord,"==",each_similarity)
        print("==================================================")
    print("list similarity====",similarity_lst,"Length==",len(similarity_lst)) 
    #Replace 0 instead of None in the list because None is not okay to calculate min and max function
    for similar_index,similarity_value in enumerate(similarity_lst):
        if similarity_value == None:
            similarity_lst[similar_index]=0
    print("modified similarity====",similarity_lst,"Length==",len(similarity_lst)) 
    min_value=min(similarity_lst)
    max_value=max(similarity_lst)
    min_value_lst.append(min_value)
    max_value_lst.append(max_value)
    print("maximum value=====",max_value)
    print("minimum value=====",min_value)    
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
print("min lst===",min_value_lst)
print("max lst===",max_value_lst)

#For Final Result

for index,first_word in enumerate(first_column_lst):
    print("Path similarity min and max vlaues of ",first_word," and ",second_column_clean_lst[index],"====",max_value_lst[index]," and ",min_value_lst[index])

    
    
```
